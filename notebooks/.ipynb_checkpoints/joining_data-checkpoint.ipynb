Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
# imports
import requests
import pandas as pd
import json
import os
import seaborn as sns
import matplotlib.pyplot as plt
import sqlite3
```

```python
InitialFourSquareDB = pd.read_csv('four_square_venues.csv')
FourSquareDB = InitialFourSquareDB[InitialFourSquareDB['Distance'] <= 500] #Only a few extreme outliers are dropped
FourSquareDB.insert(loc = 3, column = 'Ratings', value = None)
FourSquareDB.insert(loc = 4, column = 'Review Count', value = None) #This DataFrame is standardized to allow easy merging with the YelpFusion DataFrame.
FourSquareDB.insert(loc = 6, column = 'Source', value = 'FourSquare') #A new column is added to identify 
FourSquareDB
```

```python
InitialYelpFusionDB = pd.read_csv('yelp_fusion_venues.csv')
YelpFusionDB = InitialYelpFusionDB[InitialYelpFusionDB['Distance'] <= 500] #Around 500 values end up getting dropped.
YelpFusionDB.insert(loc = 5, column = 'Source', value = 'Yelp')
YelpFusionDB
```

```python
BikesDB = pd.read_csv('bike_station_output.csv') #All of the DataFrames are stored as CSVs, and are easily retrieved.
BikesDB
```

```python
combineddata = pd.concat(
    objs = [YelpFusionDB, FourSquareDB], #The FourSquare and Yelp data is combined into a single DataFrame
    axis=0,
    join='outer',
    ignore_index = True
    
)
combineddata
```

```python
InitialFinalDB = pd.merge( #The two remaining DataFrames are merged into one. It took some tinkering to get this one working but there don't appear to be any glaring issues.
BikesDB,
combineddata,
how="inner",
on=['Station Name']
)
InitialFinalDB.sort_values(['Venue Name', 'Station Name']).head(50) #Used variations of this code to search for duplicates and get a feel for the final DataFrame.

FinalDB = InitialFinalDB.drop_duplicates(subset = ['Venue Name', 'Station Name', 'Distance']) #No rows were dropped using this criteria.

FinalDB.to_csv('ModellingDB.csv', index = False)

FinalDB.sort_values('Venue Name') #Some duplicate venues may still appear, but it is practically impossible to filter further based on the limited data provided by Yelp and FourSquare. 
#There are several Franchises that would be incorrectly wiped out if I were to filter only by Venue Name and Station Name.
```

Provide a visualization that you used as part of your EDA process. Explain the initial pattern or relationship you discoved through this visualization. 

```python
sns.histplot(data = FinalDB, x='Ratings', stat = 'probability') #Low ratings are extremely uncommon on Yelp. Probably not ideal.
```

```python
sns.histplot(data = FinalDB, x='Distance', hue='Source') #FourSquare seems to heavily favour venues that are nearer to the coordinates provided. Yelp has a much more stable distribution.
```

```python
sns.histplot(data = FinalDB, x='Slots') #Very few stations have over 40 or under 10 bike slots.
```
